#### Filtrando DataFrames

### *.LOC*

* Outra maneira de acessar uma coluna do DataFrame é utilizando o comando .loc cuja propriedade é acessar um subconjunto de linhas e/ou colunas considerando seus respectivos rótulos. Esse comando lembra bastante o fatiamento (slicing) de istas e strings que utilizamos em Design de. Software

```python
#Para Filtrar Dataframes é interessante criar variaveis com sentidos
serie_region = dados["region"] #Esta variavel que denota regiões
```

É muito importante deixar claro que após um Dataframe ser utilizado para ter partes de seus dados separados, como foi o caso da separação feita por regiões, a variavel que receber está separação não será mais considerada nem tratada como um DataFrame, mas sim como uma Serie

```python
#Para Checar o que foi dito acima se usa Type, função que indica o tipo do arquivo ou variavel
print(type(dados))
print(type(serie_region))
#Concluindo, a Serie criada teve uma alteração em sua Classe
```

Agora para checar e mostrar que a separação e translação de parte do DataFrame para a Serie funcionou:

```python
serie_region
```

### *.ILOC*

* Para Acessar o DataFrame tambem é possível utilizar o iloc cuja propriedade é acessar um subconjunto de linhas e/ou colunas considerando números interios para indicar a posição. *Algo importante de lembrar é que a posição começa em 0 e termina em lenght -1*

* *Função Geral*

`DataFrame.iloc[<posição linha(s)>,<posição coluna(s)>]`

```python
#Devemos, primeiramente, checar a quantidade de informações que temos em dada coluna
#Neste caso da de Regiões
list(dados).index("region")
```

```python
dados.iloc[:,4].head(5) #Com esse comando estou pedindo para que iloc enumere quais os diferentes tópicos
```

#### *Loc vs iLoc*

```python
dados_ordenados_por_populacao
```

```python
dados_ordenados_por_populacao.loc[0:2,:]
#Função Mostra todos os paises com população de forma crescente
```

```python
dados_ordenados_por_populacao.iloc[0:2,:]
#Função mostra apenas o Maior pais de certa região
```

### Selecionando Linhas

* Sem Condicional

Antes a primeira definição de Loc e iLoc estava preenchida com :. Digamos agora que eu queria filtrar o `DataFrame`deixando apenas um grupo selecionado de paises. Tudo que vou precisar fazer é utilizar o primeiro argumento para selecionar algumas linhas específicas

```python
print("Regiões:")
print(dados["region"])
print("\n\nComparação das regiões iguais à string 'america_south':")
print(dados["region"] == "america_south")
#Essa Função vai imprimir diversos valores booleanos que poderão ser utilizados para saber em qual das linhas tem os dados desejados, neste cado da America_south
```

```python
#Filtra_Linhas vai definir quais paises pertencem a região da America_South
filtra_linhas = dados["region"] == "america_south"
#Atraves do Loc Vamos pedir para criar um DataFrame com Apenas os paises Filtrados
dados.loc[filtra_linhas,:]
```

* Com Condicional

```python
america_do_sul = dados["region"] == "america_south"
populacao_menor_que_10m = dados["Population"]<10000000
dados.loc[america_do_sul & populacao_menor_que_10m,:]
#Essa Função não pede apenas uma região demarcada, mas tambem é limitada pela população
#Maior que 10 Milhões de habitantes
```

#### *Uso Simplificado*

A Função Abaixo têm as mesmas funcionalidades das funções acima, só que ela consegue replicá-las de forma enxuta

```python
dados.loc[(dados["region"] == "america_south") & (dados["Population"]<10000000),:]
```

### *Selecionando Subconjuntos*

## Resumo dos Comandos

Aqui você encontra um resumo dos comandos apresentados neste notebook:

- [`read_excel`](#read_excel): abre um arquivo do tipo XLSX;
- [`head`](#head): mostra apenas as primeiras linhas do `DataFrame`;
- [`tail`](#tail): mostra apenas as últimas linhas do `DataFrame`;
- [`sort_values`](#sort_values): devolve um **novo** `DataFrame` com o conteúdo ordenado;
- [`loc`](#loc): acessa um subconjunto de linhas e/ou colunas considerando seus respectivos rótulos no DataFrame.
- [`iloc`](#iloc): acessa um subconjunto de linhas e/ou colunas considerando números inteiros para indicar a posição.
