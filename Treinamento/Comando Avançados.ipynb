**BASE DE DADOS e DICIONÁRIO DAS VARIÁVEIS**

A base de dados foi extraída da plataforma [Kaggle](https://www.kaggle.com) intitulado “Analyze diamonds by their cut, color, clarity, price, and other atributes”, com adaptações de textos e dos dados para fins didáticos. Ainda, a base de dados original passou por algumas limpezas e transformações.

Essa base de dados possui os preços e outros atributos de dois mil diamantes.

As variáveis apresentadas na base de dados `diamonds.xlsx` são:
 * `id`: número de identificação do diamante
 * `preco`: preço do diamante (em milhares de dólares);
 * `corte`: qualidade do corte com as categorias ordinais: 'Moderado', 'Bom', 'Muito bom', 'Premium', 'Ideal';
 * `cor`: cor do diamante com as categorias ordinais: 'D' (pior) até 'J' (melhor);
 * `claridade`: indica quão claro é o diamante com as categorias ordinais: 'I1' (pior), 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1' e 'IF' (melhor);
 * `peso`: peso do diamante (em gramas);
 * `comprimento`: comprimento do diamante (em milímetros);
 * `largura`: largura do diamante (em milímetros); e
 * `profundidade`: profundidade do diamante (em milímetros).

```python
#Carregando Dados Utilizados
dados = pd.read_excel("diamonds.xlsx")
#Checando Dados
dados.head(3)
```

```python
#Categorizando DataFrame
dados["corte"] = dados["corte"].astype("category")
dados["cor"] = dados["cor"].astype("category")
dados["claridade"] = dados["claridade"].astype("category")
#Definindo Qualidade de Corte
```

Considerando o *dataframe* completo, faça:

```python
#Definindo Scatter + Cor + Transparência
plt.scatter(dados["peso"],dados["preco"],c="red",alpha=0.4)
plt.title("Gráfico de Dispersão - Peso vs. Preço")
plt.ylabel("Preço")
plt.xlabel("Peso")
plt.show()
```

```python
#Definindo Covariancia
cov = dados["peso"].cov(dados["preco"])
#Definindo Correlação Entre Peso e Preço
correlation = dados["peso"].corr(dados["preco"])
print(f"A Correlação entre o Peso e Preço de um Diamante é de: ={correlation:.3f} e a Covariancia é de: ={cov:.3f}")
```

*Analisando os dados do gráfico e a conta feita da razão entre peso e preço é notório que a medida que o valores vão aumentando o peso também aumenta, sendo válida a ideia tida do senso comum de que quanto maior o peso maior o preço. Vale ressaltar ainda que o valor não se assemelha tanto a uma reta não fazendo esse dito ser totalmente verdadeiro mas podemos dizer que se trata de uma associação positiva e moderada pra forte

  * No *dataframe* `df2`, crie a variável `volume` contendo o volume aproximado do diamante a partir das três medições do diamante: `comprimento`, `largura` e `profundidade`. 
  * Organize as linhas do *dataframe* `df2` de tal forma que a primeira linha contenha o diamante com maior volume (em milímetro cúbico).
  * Visualize o conteúdo do *dataframe* com a ordenação feita acima e responda qual o `id` da primeira linha.

```python
#Grupo de Cores
cor1 = dados["cor"]== "J"
cor2 = dados["cor"]=="H"
#Separando DataFrame df2 nas cores conseguidas
df2 = dados.loc[cor1|cor2,:]
#Checando dataframe
df2.head(3)
```

```python
df2 = df2.copy()
#Criando Coluna Volume
df2["volume"] = df2["comprimento"]*df2["largura"]*df2["profundidade"]
#Deixando Apenas Coluna Volume
df3 = df2.loc[:,["volume"]]
df3.sort_values(["volume"], ascending=False)
```
